```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x-values
x = np.linspace(0, 2 * np.pi, 400)

# Compute sin and cos
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label="sin(x)")
plt.plot(x, y_cos, label="cos(x)")
plt.title("Basic Sine and Cosine Waves")
plt.xlabel("x (radians)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```
