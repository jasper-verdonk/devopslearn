```python
import pandas as pd
import ipaddress
import re
```

```python
def ip_in_prefix_string(ip, prefix_string):
    ip_obj = ipaddress.ip_address(ip)

    # Split prefixes by comma, semicolon, or whitespace
    prefixes = re.split(r'[,\s;]+', prefix_string.strip())

    for p in prefixes:
        if not p:
            continue   # if p is False, skip the iteration but continue with the next iteration.
        try:
            network = ipaddress.ip_network(p, strict=False)
            if ip_obj in network:
                return True
        except ValueError:
            # skip malformed prefixes
            continue

    return False
```
