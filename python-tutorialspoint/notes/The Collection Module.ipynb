## DefaultDict

Sub class of the dictionary

```python
from collections import defaultdict
```

```python
dd = defaultdict(int)
```

```python
nums = [1, 2, 3, 4, 2, 4, 1, 2]
```

```python
for i in nums:
    dd[i] += 1
```

```python
print(dd)
```

```python
print(dd[1])
```

```python
print(dd[5])
```

default value is set to zero of keys that are not defined without raising an error. 

```python
dd2 = defaultdict(list)
```

```python
for i in nums:
    dd2[i].append(i)
```

```python
print(dd2)
```

```python
print(dd2[2])
```

```python
print(dd2[5])
```

## ChainMap

```python
from collections import ChainMap
```

```python
d1 = {'a':1, 'b':2}
d2= {'c':3, 'd':4}
d3= {'e':5, 'f':6}
```

```python
c = ChainMap(d1, d2, d3)
```

```python
print(c)
```

```python
print(c['a'])
print("---")
for i in c.values():
    print(i)
```

```python
for i in c.keys():
    print(i)
```

```python
for a, b in c.items():
    print(a)
    print(b)
```

```python
d4 = {'x':8, 'y':9}
```

```python
c1 = c.new_child(d4)
```

```python
print(c1)
```

Dictionary's are unorded in nature so the 'order' is at random. 

## Named Tuple

```python
my_tuple = ('Dave', 'Londen', 40)
```

```python
print("Lives in: ", my_tuple[1])
```

```python
from collections import namedtuple
```

```python
Employee = namedtuple('Employee', ['name', 'location', 'age'])
```

```python
emp1 = Employee('Dave', 'Londen', 40)
```

```python
print("The employee is", emp1.age)
```

```python
print("The employee name is:", emp1.name)
```
