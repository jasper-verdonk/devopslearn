Step 1. Install ipykernel in your virtual environment

Make sure you have the Jupyter kernel package inside your virtual environment:

pip install ipykernel

Step 2. Add your venv as a Jupyter kernel

Still inside your virtual environment, run:

python -m ipykernel install --user --name=dash_env --display-name "Python (dash_env)"


This registers your virtual environment (dash_env) as a selectable kernel in Jupyter.

```python
import dash
from dash import html
from dash import dcc
```

2. Application Instantiation

```python
app = dash.Dash(__name__)
```

3. App Layout

```python
app.layout = html.Div([
    html.H1("My First Dash App"),
    dcc.Dropdown(
        options=[
            {'label': 'Option 1', 'value': '1'},
            {'label': 'Option 2', 'value': '2'},
        ],
        value='1',
        id='dropdown-example'
    ),
    dcc.Slider(0, 10, 1, value=5, id='slider-example'),
    dcc.Graph(id='graph-example')
])
```

4. Callback functions

5. Running the app

```python
if __name__ == '__main__':
    app.run(debug=True)
```
