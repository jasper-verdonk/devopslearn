# Reading from .csv files

```python
import csv
```

```python
with open('csv.txt' , 'r') as file:

    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
FILE_PATH = 'csv.txt' 

class Hobby:
    HAS_HEADERS = True
    NAME_COL = 0
    AGE_COL = 1
    FAV_HOBBY_COL = 2

with open(FILE_PATH, 'r') as file:

    reader = csv.reader(file)
    headers = []
    if Hobby.HAS_HEADERS:
        headers = next(reader)
    for row in reader:
        if row and len(row) > Hobby.FAV_HOBBY_COL:
            name = row[Hobby.NAME_COL]
            age = int(row[Hobby.AGE_COL])

            if age >= 20 and age <= 30:
                print('{} is {} years old.'.format(name, age))
            
            
```

# Writing to .csv files

```python
import csv

headers = ['student_name', 'age', 'fav_subject']

rows = [
    ['Harry', 20, 'Maths'],
    ['Kim', 27, 'English'],
    ['Dave', 36, 'Biology']
]

with open('school_csv', 'w', newline='') as file:

    writer = csv.writer(file)
    writer.writerow(headers)
    writer.writerows(rows)

    
```

```python
mydict = [
    {'region': 'England', 'name': 'Fruit Store', 'sales': 1000 },
    {'region': 'Scotland', 'name': 'Electronics', 'sales': 50000 },
    {'region': 'England', 'name': 'Card Store', 'sales': 5000 },
    {'region': 'Ireland', 'name': 'Sports Store', 'sales': 15000 }
]

fields = ['region', 'name', 'sales']

with open('shop_csv', 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames = fields)
    writer.writeheader()
    writer.writerows(mydict)
```
