```python
import numpy as np
import matplotlib.pyplot as plt

def polygon_perimeter(n, r=1):
    # side length for n-gon with radius r
    side = 2 * r * np.sin(np.pi / n)
    return n * side

r = 1
nsides = [6, 12, 24, 48, 96, 192]
perimeters = [polygon_perimeter(n, r) for n in nsides]

plt.figure(figsize=(8, 5))
plt.plot(nsides, perimeters, 'o-', label="Polygon Perimeter")
plt.axhline(2*np.pi*r, color='grey', linestyle='--', label="Circle Circumference = 2Ï€r")
plt.title("Polygon Perimeter Approaching Circle Circumference")
plt.xlabel("Number of polygon sides")
plt.ylabel("Perimeter")
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

radii = np.linspace(0.5, 5, 50)
diameters = 2 * radii
circumferences = 2 * np.pi * radii

plt.figure(figsize=(8, 5))
plt.plot(diameters, circumferences, label="C vs d")
plt.title("Circumference is Proportional to Diameter")
plt.xlabel("Diameter")
plt.ylabel("Circumference")
plt.grid()
plt.legend()
plt.show()
```
