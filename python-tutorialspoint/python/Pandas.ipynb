```python
import pandas as pd
from fpdf import FPDF
```

```python
df = pd.DataFrame({
    "Location": location, 
    "Short desc": short_desc,
    "Subnet": subnet,
    "VRF name": vrf_name,
    "GW IP": gw_ip,
    "Subnet mask": subnet_mask,
    "Host IP": host_ip
})
```

```python
location = ['MGW113', 'MGW113', 'MGW113', 'MGW113', 'CAR721', 'CAR721', 'CAR721', 'CAR721']
short_desc = ['MGMT', 'ILO', 'Unicast Contri/Distri', 'Mulicast Contri/Distri', 'MGMT', 'ILO', 'Unicast Contri/Distri', 'Multicat Contri/Distri']
vrf_name = ['onm_sonm_csstr', 'onm_sonm_cssrt_oob', 'exs_cexs_srt', 'acs_macs', 'onm_sonm_csstr', 'onm_sonm_cssrt_oob', 'exs_cexs_srt', 'acs_macs']
subnet = ['172.20.19.160/28', '172.20.19.176/28', '172.23.45.32/28', '172.23.44.0/26', '172.20.19.192/28', '172.20.19.208/28', '172.23.45.48/28', '172.23.44.128/26']
gw_ip = ['172.20.19.161', '172.20.19.177', '172.23.45.33', '172.23.44.1', '172.20.19.193', '172.20.19.209', '172.23.45.49', '172.23.44.129']
subnet_mask = ['255.255.255.240', '255.255.255.240', '255.255.255.240', '255.255.255.192', '255.255.255.240', '255.255.255.240', '255.255.255.240', '255.255.255.192']
host_ip = ['172.20.19.174', '172.20.19.190', '172.23.45.46', '172.23.44.20', '172.20.19.206', '172.20.19.212', '172.23.45.62', '172.23.44.157']
```

```python
pdf = FPDF()
pdf.add_page(orientation="L")
pdf.set_font("Helvetica", size=12)
```

```python
col_widths = [20, 40, 40, 60, 40, 40, 40, 40]
```

```python
# Add header row
for col_name, width in zip(df.columns, col_widths):
    pdf.cell(width, 10, col_name, border=1, align="C")
pdf.ln(10)

# Add data rows
for _, row in df.iterrows():
    for item, width in zip(row, col_widths):
        pdf.cell(width, 10, str(item), border=1, align="C")
    pdf.ln(10)
```

```python
pdf.output("pdf IP allocation SRT.pdf")
```
