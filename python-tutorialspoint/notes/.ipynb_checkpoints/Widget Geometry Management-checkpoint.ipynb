```python
root = Tk()

menu = Menu(root)

root.config(menu=menu)

filemenu = Menu(menu)
menu.add_cascade(label="File", menu=filemenu)
filemenu.add_command(label= "New")
filemenu.add_command(label= "Open")
filemenu.add_command(label= "Save")
filemenu.add_separator()
filemenu.add_command(label = "Exit", command=root.destroy)

helpmenu = Menu(menu)
menu.add_cascade(label="Help", menu=helpmenu)
helpmenu.add_command(label = "Contact")

root.mainloop()
```

## Executing functions in a GUI

```python
from tkinter import *
from tkinter import messagebox

root = Tk()
root.geometry('200x100')

def showMsg():
    messagebox.showinfo('Message', 'Nicely Clicked!')

button = Button(root, text = 'Click', command = showMsg)
button.pack()

root.mainloop()
```

functions must be created before the button definition. 

```python
from tkinter import * 

root = Tk()
root.geometry('200x100')

text_Input = StringVar()
total = '0'

def btnClick():
    global total
    total = int(total) + 1
    text_Input.set(str(total))

txtDisplay = Entry(root, justify='right', textvariable = text_Input).pack()

counter = Button(root, text = 'Count', command = btnClick).pack()

root.mainloop()
```
