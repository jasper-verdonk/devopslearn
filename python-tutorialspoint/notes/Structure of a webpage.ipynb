# National Weather Services - Extended Forecast

````python
page = requests.get("https://forecast.weather.gov/MapClick.php?lat=34.052238&lon=-118.243344")
soup = BeautifulSoup(page.content, 'html.parser')
````

````python
page
````

````python
seven_day_forcast = soup.find(id="seven-day-forecast")
forecast_items = seven_day_forcast.find_all(class_="tombstone-container")
````

````python
seven_day_forcast
````

````python
forecast_items
````

````python
today = forecast_items[0]
````

````python
print(today.prettify())
````

````python
type(forecast_items)
````

````python
period = today.find(class_="period-name").get_text()
````

````python
print(period)
````

````python
short_desc = today.find(class_="short-desc").get_text()
````

````python
print(short_desc)
````

````python
temp = today.find(class_="temp").get_text()
````

````python
print(temp)
````

````python
img = today.find("img")
desc = img['title']
print(desc)
````

````python
img
````

````python
desc_2 = img['class']
print(desc_2)
````

Beautiful soup objects support searching a page via CSS selectors. We can use the select method. 
```
https://www.w3schools.com/cssref/css_selectors.php
```

````python
period_tags = seven_day_forcast.select(".tombstone-container .period-name")
````

````python
period_tags
````

````python
periods = [pt.get_text() for pt in period_tags]
````

````python
periods
````

````python
short_descs = [sd.get_text() for sd in seven_day_forcast.select(".tombstone-container .short-desc")]
````

````python
short_descs
````

````python
temps = [t.get_text() for t in seven_day_forcast.select(".tombstone-container .temp")]
````

````python
temps
````

````python
descs = [d['title'] for d in seven_day_forcast.select(".tombstone-container img")]
````

````python
descs
````

## Loading to a Pandas DataFrame

````python
import pandas as pd
````

````python
weather = pd.DataFrame({
    "period": periods, 
    "short_desc": short_descs,
    "temp": temps, 
    "desc": descs
})
````

````python
weather
````

````python
search = "(\d+)"
temp_nums = weather["temp"].str.extract(search, expand=False)
weather["temp_num"] = temp_nums.astype('int')
````

````python
temp_nums
````

````python
weather["temp_num"].mean()
````

````python
is_low = weather["temp"].str.contains("Low")
weather["is_low"] = is_low
weather = weather[weather['is_low']] 
````
